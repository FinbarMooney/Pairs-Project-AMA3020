```python
import numpy as np
import matplotlib.pyplot as plt
m = 1.5
R = 0.121
R2 = 0.112
R3 = 0.131
I = (2/5)*m*(R**2)
I2 = (2/5)*m*(R2**2)
I3 = (2/5)*m*(R3**2)
vc = 10
g = 9.81
ac = 0.0008

ro = (I*vc**2)/(m*g*ac*R)
v = np.linspace(0,vc,100)
rv1 = (I*v**2)/(m*g*ac*R)
rvx1= rv1/100
rv2 = (I2*v**2)/(m*g*ac*R2)
rvx2= rv2/100
rv3 = (I3*v**2)/(m*g*ac*R3)
rvx3= rv3/100
plt.figure(figsize=(8,6))
plt.plot(v,rvx1)
plt.plot(v,rvx2)
plt.plot(v,rvx3)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Velocity (m/s)',fontsize=16)
plt.ylabel(r'$\rho\cdot 10^2$ (m)',fontsize=16)
plt.legend(['R = 0.121 m','R = 0.112 m','R = 0.131 m'], fontsize=16)
plt.savefig('Velocity.pdf')
plt.show()


a = np.linspace(0.0001,0.01,100)
ra1 = (I*vc**2)/(m*g*a*R)
rax1 = ra1/1000
ra2 = (I2*vc**2)/(m*g*a*R2)
rax2 = ra2/1000
ra3 = (I3*vc**2)/(m*g*a*R3)
rax3 = ra3/1000

plt.figure(figsize=(8,6))
plt.plot(a,rax1)
plt.plot(a,rax2)
plt.plot(a,rax3)
plt.legend(['R = 0.121 m','R = 0.112m','R = 0.131 m'], fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Eccentricity (m)',fontsize = 16)
plt.ylabel(r'$\rho\cdot 10^3$ (m)',fontsize=16)
plt.savefig('Eccentricity.pdf')
plt.show()
```
